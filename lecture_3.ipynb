```python
df['Form'].fillna(-1, inplace = True)
```

```python
df['Form'] = df['Form'].astype(int)
```

```python
df['Form'] = df['Form'].astype(str)
```

```python
df.Form.value_counts().plot(kind='bar')
```

```python
val_count = df.Form.value_counts()
```

```python
val_count[val_count.values > 10].plot(kind = 'bar')
```

```python
val_count = df[df.Form != '-1'].Form.value_counts()

val_count[val_count.values > 10].plot(kind = 'bar')
```

```python
df['Form'].value_counts().plot(kind='bar', label='Churn')
plt.legend()
plt.title('Распределение оттока клиентов');
```
